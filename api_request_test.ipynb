**Setting up an API call to get historical financial data**  
Connect to 12finance free API to request historical stock data for a stock of my choosing

Plan:
- set up a get request 
    - import requests library
- send request to 12finance API
- recieve data back in JSON format and manipulate into pandas DF
    - extrac

```python
import requests
import pandas as pd
import credentials

#api endpoint 
URL = "https://api.twelvedata.com/time_series?"

#api parameters **EDIT THESE**
api_key = credentials.twelve_data_api_key
interval = '1day'
symbol = 'TSLA'
outputsize = '5000'

#create dictionary of parameters
PARAMS = {'apikey': api_key, 'interval':interval, 'symbol':symbol, 'outputsize':outputsize}

#send get request and save response as response object
r = requests.get(url = URL, params = PARAMS)

#extract data in json format
data = r.json()

#extract time series data
time_series = data['values']



stock = pd.DataFrame(time_series)

stock.head(50)
```
